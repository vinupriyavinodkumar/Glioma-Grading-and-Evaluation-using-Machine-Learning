## Checking the prevalence of Race on Gene Mutations

```python
import pandas as pd
import matplotlib.pyplot as plt


# Explore the data
race_counts = input_df['Race'].value_counts()
print("Race Counts:\n", race_counts)

# Visualize gene mutations by race
plt.figure(figsize=(15, 10))
for gene_column in input_df.columns[5:]:
    gene_by_race = input_df.groupby('Race')[gene_column].mean()
    plt.bar(gene_by_race.index, gene_by_race, label=gene_column)

plt.xlabel('Race')
plt.ylabel('Mean Gene Mutation')
plt.title('Gene Mutations by Race')
plt.legend()
plt.show()

# Perform statistical analysis (e.g., ANOVA or Chi-squared test)
# You can use libraries like scipy or statsmodels for this.
```

```python
# Calculate the sum of gene mutations by race
total_mutations_by_race = input_df.groupby('Race').sum().sum(axis=1)

# Calculate the total samples by race
total_samples_by_race = input_df['Race'].value_counts()

# Calculate the proportion of gene mutations by race
proportion_by_race = total_mutations_by_race / total_samples_by_race

# Print the proportion
print("Proportion of Gene Mutations by Race:\n", proportion_by_race)

# Visualize the proportion
plt.figure(figsize=(8, 5))
plt.bar(proportion_by_race.index, proportion_by_race)
plt.xlabel('Race')
plt.ylabel('Proportion of Gene Mutations')
plt.title('Proportion of Gene Mutations by Race')
plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming input_df is your DataFrame containing race and gene mutation data
# Remove the rows where Race is 'American Indian or Alaska Native'
filtered_df = input_df[input_df['Race'] != 'american indian or alaska native']

# Sum the number of mutations for each gene by race
sum_by_race = filtered_df.groupby('Race').sum()

# Chi-Squared Test
chi2, p_value, dof, expected = chi2_contingency(sum_by_race)

print("Chi2:", chi2)
print("P-Value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:", expected)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("There is a statistically significant relationship between race and gene mutations.")
else:
    print("There is no statistically significant relationship between race and gene mutations.")
```

## Checking Gender Affinity towards Gene Mutations

```python
import matplotlib.pyplot as plt

gender_column = 'Gender'

# List of gene columns
gene_columns = ['IDH1', 'TP53', 'ATRX', 'PTEN', 'EGFR', 
                'CIC', 'MUC16', 'PIK3CA', 'NF1', 'PIK3R1', 
                'FUBP1', 'RB1', 'NOTCH1', 'BCOR', 'CSMD3', 
                'SMARCA4', 'GRIN2A', 'IDH2', 'FAT4', 'PDGFRA']
# Calculate the total number of mutated genes for each individual
input_df['Total_Mutations'] = input_df[gene_columns].apply(lambda row: row.eq(1).sum(), axis=1)

# Sum of total mutations for each gender
sum_mutations_by_gender = input_df.groupby(gender_column)['Total_Mutations'].sum()

# Total mutations across all individuals
total_mutations_all = input_df['Total_Mutations'].sum()

# Calculate the percentage of mutations for each gender
percentage_mutations_by_gender = (sum_mutations_by_gender / total_mutations_all) * 100

# Labels for the chart
labels = ['Male', 'Female']

# Plotting
plt.figure(figsize=(5, 5))
plt.pie(percentage_mutations_by_gender, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Total Gene Mutations by Gender')
plt.show()

# Replace numerical identifiers with string equivalents for clearer output
#percentage_mutations_by_gender.index = percentage_mutations_by_gender.index.map(gender_map)

print("Percentage of mutations by gender:")
print(percentage_mutations_by_gender)
```

```python
from scipy.stats import chi2_contingency
import numpy as np

# Creating a contingency table
# Gender 0 is male, Gender 1 is female
male_mutated = input_df[(input_df['Gender'] == 0) & (input_df['Total_Mutations'] > 0)].shape[0]
male_not_mutated = input_df[(input_df['Gender'] == 0) & (input_df['Total_Mutations'] == 0)].shape[0]
female_mutated = input_df[(input_df['Gender'] == 1) & (input_df['Total_Mutations'] > 0)].shape[0]
female_not_mutated = input_df[(input_df['Gender'] == 1) & (input_df['Total_Mutations'] == 0)].shape[0]

contingency_table = np.array([[male_mutated, male_not_mutated], 
                              [female_mutated, female_not_mutated]])

# Performing the Chi-squared test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2: {chi2}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: {expected}")

if p_value < 0.05:
    print("There is a statistically significant relationship between gender and gene mutations.")
else:
    print("There is no statistically significant relationship between gender and gene mutations.")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```
